```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',flavor = 'html5lib')
cf=pd.DataFrame(df[0])
cf.head()
```

```python
cf.rename({0:'Postcode',1:'Borough',2:'Neighbourhood'},axis='columns',inplace=True)
cf.head()
```

```python
print(cf.isnull().sum())
cf.drop(cf.index[0],inplace=True)
cf.dropna(axis=0,inplace=True)
cf.head()
print(cf.isnull().sum())
cf
```

```python
print('the df has {} boroughs and {} neighbours'.format(
len(cf['Borough'].unique()),
len(cf['Neighbourhood'].unique())))
```

```python
address = 'Toronto, Ontario, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)
for lat,lng,label in zip(new_cf['Latitude'],new_cf['Longitude'],new_cf['Neighbourhood']):
    label=folium.Popup(label,parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='red',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False
    ).add_to(map_toronto)
```
